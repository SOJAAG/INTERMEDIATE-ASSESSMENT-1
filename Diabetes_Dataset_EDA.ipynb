```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Univariate Analysis


```python
data=pd.read_csv('diabetes_model.csv')
data
```

Q1.plot histogram for the above data

```python
plt.hist(data["Insulin"],color="blue",rwidth=0.8,density=True)
plt.title("Histogram of Diabetes Dataset")
plt.xlabel("Insulin")
plt.ylabel("Density")
plt.show()
```

Q2.plot Boxplot of Glucose from Diabetes dataset

```python
import seaborn as sns
```

```python
np.mean(data["Glucose"])
```

```python
np.std(data["Glucose"])
```

```python

sns.boxplot(data["Glucose"])


plt.title("Box Plot of Dataset ")
plt.show()
```

Bivariate Analysis


Q3.Scatter plot of Blood Pressure vs BMI

```python
plt.figure(figsize=(5,3))

plt.scatter(data["BloodPressure"],data["BMI"], c="b", marker="*")

plt.title("Blood Pressure vs BMI")
plt.xlabel("BloodPressure")
plt.ylabel("BMI")
plt.show()
```

Q4.Correlation matrix for Diabetes

```python


correlation_matrix = data.corr()

print("Correlation Matrix:\n",correlation_matrix)
```

Multivariate Analysis

Q5.Pair plot of Diabetes dataset

```python
sns.pairplot(data,hue="Outcome", x_vars=["Glucose"])
plt.show()
```

Q6. 3D Plot for the Dataset Diabetes

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.scatter(data["Glucose"],data["BloodPressure"],data["BMI"],c=data['Outcome'],cmap='viridis')
ax.set_xlabel("Glucose")
ax.set_ylabel("BloodPressure")
ax.set_zlabel("BMI")
plt.title("3D Plot of Diabetes Dataset")
plt.show()
```
